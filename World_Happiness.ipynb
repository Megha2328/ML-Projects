```python
# Ladder Score Distribution by Regional Indicator
plt.figure(figsize=(15,8))
# kde-kernel density estimate- used to plot probability density function of continuous or
# non-parametric data, where we can plot the data for various samples in a single graph for 
# better visiualization
sns.kdeplot(x=df_2021["Ladder score"],hue=df_2021["Regional indicator"],fill=True,linewidth=2)
plt.axvline(x=df_2021["Ladder score"].mean(),color="black") # to add an infinitely long line-here the black line
plt.title("Ladder Score Distribution by Regional Indicator")
plt.show()
```

```python
all_countries=df["Country name"].unique()
all_countries_pop=pg["Country Name"].unique()

del_count=[]
for i in all_countries:
    if i not in all_countries_pop:
        del_count.append(i)
del_count
```

```python
df.columns
```

```python
pop_df = df[['Log GDP per capita', 'Life Ladder', 'Country name', 'year', 'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity',"region",'Perceptions of corruption']].copy()
pop_df.head()
```

```python
pop_df=pop_df[~pop_df["Country name"].isin(del_count)]
pop_df=pop_df[~pop_df.year.isin([2005,2006,2007,2018,2019,2020,2021])]
pop_dict={i:{} for i in range(2008,2018)}
for i in range(len(pg)):
    if(pg["Year"][i] in range(2008,2018)):
        pop_dict[pg["Year"][i]][pg["Country Name"][i]]=pg["Count"][i]
```

```python
population=[]
for i in pop_df.index:
    population.append(pop_dict[pop_df["year"][i]][pop_df["Country name"][i]])
pop_df["population"]=population
```

```python
pop_df.head()
```

```python
fig=px.scatter(pop_df,
               x="Log GDP per capita",
               y="Life Ladder",
               animation_frame="year",
               animation_group="Country name",
               size="population",
               template="plotly_white",
               color="region",
               hover_name="Country name",
               size_max=60)
fig.update_layout(title="Life Ladder and Log GDP per capita Comparison by Countries via Regions for each Year")
fig.show()
```
